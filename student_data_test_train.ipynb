```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("student_info.csv")
df
```

```python
df.shape
```

```python
plt.scatter(x=df.study_hours,y=df.student_marks,color='g')
plt.title('Student Data Visualization')
plt.xlabel('Study Hours')
plt.ylabel('Student Marks')
plt.show()
```

```python
df.isnull().sum()
```

```python
df.mean()
```

```python
df2=df.fillna(df.mean())
df2.head()
```

```python
df2.isnull().sum()
```

```python
#separating data into training and testing
x=df2.drop('student_marks',axis=1)
y=df2.drop('study_hours',axis=1)
```

```python
x.shape
```

```python
y.shape
```

```python
#split dataset into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#test the model with given value
lr.predict([[8.5]])[0][0].round(2)
```

```python
sh=float(input("enter the study hours"))
lr.predict([[sh]])[0][0].round(2)
```

```python
lr.score(x_test,y_test)
```
