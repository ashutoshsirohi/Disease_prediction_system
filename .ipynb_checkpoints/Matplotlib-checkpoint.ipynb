```python
from numpy import *
ar1=array([
    [2,3,4],
    [9,8,0]
])
ar2=ar1.reshape(3,2)
print(ar2)
```

```python
#3D array
from numpy import *
ar1=array([
    [2,3,4,5,6,7],
    [9,8,0,1,2,3]
])
ar2=ar1.flatten()
ar3=ar2.reshape(2,2,3)
print(ar3)
```

```python
#matrix
from numpy import *
ar1=array([
    [2,3,4,5,],
    [9,8,0,1]
])
m=matrix(ar1)
print(m)
print(type(m))
```

```python
m=matrix('1,2,3;4,5,6;6,7,8')
print(m)
```

```python
#to multiply 2 matrix
m1=matrix('1,2,3;4,5,6;6,7,8')
m2=matrix('1,2,3;4,5,6;6,7,8')
m3=m1*m2
print(m3)
```

```python
m1=matrix('1,2,3;4,5,6;6,7,8')
m2=matrix('1,2,3;4,5,6;6,7,8')
m3=m1*m2
print(m3)     
```

```python
from numpy import *
ar1=array([
    [2,3,4,5,6,7],
    [9,8,0,1,2,3]
])
ar2=ar1.flatten()
print(min(ar2))
print(max(ar2))
print(sum(ar2))
print(sqrt(ar2))
print(sin(ar2))
```

```python
#create a array by taking input from the user

ar=zeros(6,int)
for i in range(6):
    num1= int(input("enter the value"))
    ar[i]=num1
print(ar.reshape(3,2))
```

```python
#matplotlib

import matplotlib.pyplot as plt
x=['A','B','C','D']
y=[2,3,4,5]
plt.bar(x,y,color='pink')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['A','B','C','D']
y=[2,3,4,5]
c=['lightgreen','green','blue','indigo']
plt.bar(x,y,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['A','B','C','D']
y=[2,3,4,5]
c=['lightgreen','green','blue','indigo']
plt.bar(x,y,color=c,edgecolor='black',linewidth=10,width=0.3)
plt.show()
```

```python
x=['toothpaste','honey','soap','facewash']
pantanjali=[95,350,80,470]
dabur=[75,300,45,360]
plt.xlabel('Products',fontsize=15)
plt.xlabel('Cost',fontsize=15)
plt.title('GREAT WHITE SHARK!',fontsize=25)
plt.bar(x,pantanjali,color='red',label='pantajali')
plt.bar(x,dabur,color='green',label='dabur')
plt.legend()
plt.show()
```

```python
#scatter plot
from matplotlib import pyplot as plt
x=[1,2,3,4]
y=[2,3,4,5]
c=['green','cyan','blue','crimson']
s=[400,300,500,700]
plt.scatter(x,y,color=c,sizes=s,marker='d')
plt.show()
```

```python
#Pie chart
import matplotlib.pyplot as plt
x=[60,90,60,90]
y=['c','c++','java','python']
c=['red','black','red','black']
e=[0.1,0.0,0.1,0.0]
plt.pie(x,labels=y,colors=c,explode=e,autopct='%0.1f%%',shadow=True)
plt.show()
```

```python
#donate chart or ring chart
import matplotlib.pyplot as plt
x=[60,90,60,90]
y=['c','c++','java','python']
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors='white')
plt.show()
```

```python
#stack plot or Area Plot 
import matplotlib.pyplot as plt
x=[1,2,3,4]
area1=[3,4,5,7]
area2=[8,9,7,6]
area3=[1,2,3,4]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','g','y'])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
area1=[3,4,5,7]
area2=[8,9,7,6]
area3=[1,2,3,4]
l=['area1','area2','area3']
plt.subplot(2,2,1)
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','g','y'])

x=[60,90,60,90]
y=['c','c++','java','python']
c=['red','black','red','black']
e=[0.1,0.0,0.1,0.0]
plt.subplot(2,2,2)
plt.pie(x,labels=y,colors=c,explode=e,autopct='%0.1f%%',shadow=True)

x=[1,2,3,4]
y=[2,3,4,5]
c=['green','cyan','blue','crimson']
s=[400,300,500,700]
plt.subplot(2,2,3)
plt.scatter(x,y,color=c,sizes=s,marker='d')

x=['toothpaste','honey','soap','facewash']
pantanjali=[95,350,80,470]
dabur=[75,300,45,360]
plt.subplot(2,2,4)
plt.xlabel('Products',fontsize=15)
plt.xlabel('Cost',fontsize=15)
plt.bar(x,pantanjali,color='red',label='pantajali')
plt.bar(x,dabur,color='green',label='dabur')

plt.legend()
plt.show()
```
