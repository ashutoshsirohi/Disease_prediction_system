```python
import numpy as np
import pandas as pd
dataset1=pd.read_csv('sports.csv')
dataset1
```

```python
#separate dataset into dependent and independent varible
x= dataset1.iloc[:,[1,2]].values
y= dataset1.iloc[:,3].values
```

```python
x          #input values
```

```python
y          #output values
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
l1          = LabelEncoder()
x[:,1]      = l1.fit_transform(x[:,1])
```

```python
x
```

```python
y       = l1.fit_transform(y)
```

```python
y
```

```python
#Apply KNN algorithm
from sklearn.neighbors import KNeighborsClassifier
classifier1=KNeighborsClassifier(n_neighbors=3)
classifier1.fit(x,y)
```

```python
y_pred = classifier1.predict([[15,1]])
y_pred
```
